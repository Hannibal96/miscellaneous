```python
class AudioTool():
    def __init__(self, path):
        self.path = path
        self.sig, self.sr = torchaudio.load(audio_file)
        self.time_len = len(self.sig[0])
        self.fourier = np.fft.fft(np.array(self.sig[0]))
    
    def play(self):
        pygame.mixer.init()
        pygame.mixer.music.load(self.path)
        pygame.mixer.music.play()    
        
    def plot_time(self):
        plt.plot(np.arange(0, self.time_len/self.sr, 1/self.sr), self.sig[0])
        
    def plot_fourier(self):
        plt.plot(self.fourier)
    
    def LPF(self, lpf_freq=10_000):
        res_lpf = np.zeros(len(self.fourier))
        res_lpf[lpf_freq:-lpf_freq] = self.fourier[lpf_freq:-lpf_freq]
        return res_lpf
    
    def HPF(self, hpf_freq = 10_000):
        res_hpf = np.zeros(len(self.fourier))
        res_hpf[0: hpf_freq] = self.fourier[0:hpf_freq]
        res_hpf[-hpf_freq:] = self.fourier[-hpf_freq:]
        return res_hpf
```
